#** Power BI Visuals**

                                                   
Question 1 : Create a Card Visual to show Total Sales.

Ans:  The Card will display Total Sales (Sum of Sales) for the entire dataset.

Question 2  : Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the highest sales?

Ans: The ship mode generating the highest sales is usually Standard Class (in Global Superstore dataset).

Question 3 : Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any noticeable seasonal pattern in sales?

Ans: ✔ Sales generally increase year over year.
✔ Noticeable seasonal pattern: Higher sales during November–December (holiday season).

Question 4 : Create a Bar Chart to compare Sales by Category.

Ans: Typically:
Technology → Highest Sales


Furniture → Second


Office Supplies → Third
Question 5 : Create a Matrix Visual to display Sales by Month and Category and identify the month with the highest sales.

Ans:  The month with the highest sales is usually November or December.

Question 6 : Create a Decomposition Tree for Sales Breakdown.

Ans: The highest contribution generally comes from:
Technology category


Consumer segment


Certain regions like APAC or US
Question 7 : Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers where high discounts resulted in negative profit?

Ans: ✔ Yes, there are outliers.
✔ Some sub-categories (like Tables or Bookcases) show high discount but negative profit.
